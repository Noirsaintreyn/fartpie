```python
import numpy as np
import pandas as pd
import yfinance as yf
import backend
import warnings
warnings.filterwarnings('ignore')
```

```python
def predict_hod_lod_from_backend(hist_df):
    closes = hist_df["Close"]  # pandas Series
    
    # --- DAILY BACKTEST: MICROSTRUCTURE DISABLED ---
    micro = {
        "state": "Daily",
        "confidence": 1.0
    }
    
    try:
        garch = backend.calculate_garch_volatility_regime(closes.values)
        sigma_price = float((garch["current_vol"] / 100) * closes.iloc[-1])
    except (ValueError, Exception):
        # Fallback to simple historical volatility if GARCH fails
        returns = np.log(closes / closes.shift(1)).dropna() * 100
        hist_vol = returns.std()
        sigma_price = float((hist_vol / 100) * closes.iloc[-1])
    
    hod = float(closes.iloc[-1] + sigma_price)
    lod = float(closes.iloc[-1] - sigma_price)
    
    return {
        "hod": hod,
        "lod": lod,
        "sigma": sigma_price,
        "state": micro["state"],
        "confidence": micro["confidence"]
    }
```

```python
def backtest_hod_lod(ticker, start, end):
    data = yf.download(
        ticker,
        start=start,
        end=end,
        interval="1d",
        progress=False
    ).dropna()
    
    if data.empty:
        print(f"No data downloaded for {ticker}")
        return pd.DataFrame()
    
    data["date"] = data.index.date
    results = []
    
    for d, day_df in data.groupby("date"):
        hist_df = data[data.index < day_df.index[0]]
        if len(hist_df) < 300:
            continue
        
        pred = predict_hod_lod_from_backend(hist_df)
        
        hod_t = pred["hod"]
        lod_t = pred["lod"]
        sigma = pred["sigma"]
        
        # Convert to float explicitly
        hod_r = float(day_df["High"].max())
        lod_r = float(day_df["Low"].min())
        
        # Now max() will work correctly with floats
        overshoot = max(hod_r - hod_t, lod_t - lod_r, 0)
        overshoot_sigma = overshoot / (sigma + 1e-9)
        
        results.append({
            "date": d,
            "hod_theoretical": hod_t,
            "lod_theoretical": lod_t,
            "hod_realized": hod_r,
            "lod_realized": lod_r,
            "overshoot_sigma": overshoot_sigma,
            "contained": overshoot == 0,
            "state": pred["state"]
        })
    
    return pd.DataFrame(results)
```

```python
df = backtest_hod_lod(
    ticker="SPY",
    start="2012-01-01",
    end="2025-12-31"
)
```
